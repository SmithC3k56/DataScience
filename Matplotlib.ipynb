# Pyplot API vs OBject-Oriented API
- Pyplot API -> Quickly
- OO API -> Advanced

# Pyplot API

```python
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x),color='red',linestyle='dashed',linewidth=2, label='sin');
plt.plot(x, np.cos(x),color='blue',linestyle='dashed',linewidth=2, label='cos');

plt.title("Sine and Cosine")
plt.xlabel("X value")
plt.ylabel("sin of X value")
# plt.xlim([0,10])
# plt.ylim([-1,1])
#tuong tu vs xlim() va ylim() co the su dung axes()
# plt.axes([xmin,xmax,ymin,ymax])
# plt.axes("tight")
# plt.axes("equal")
plt.legend()# call ham nay moi hien label
```

# Object oriented API

```python
#1. Prepare data
x = [1,2,3,4]
y= [11,22,33,44]

#2. Setup plot
fig,ax = plt.subplots(figsize=(5,5)) #Figure size = Width & Height of the plot 

#3. Plot the data
ax.plot(x,y)
ax.set(title="A simple plot",
       xlabel="x-axis",
       ylabel="y-axis");
```

```python
# OO API

#Prepare new data

rng = np.random.RandomState(0)

x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 *rng.rand(100)

fig,ax = plt.subplots()

img1 = ax.scatter(x,y,s=sizes,c=colors,cmap='viridis',alpha=0.3);
fig.colorbar(img1);
```

# Project: Data visualization - California Cities

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
cities = pd.read_csv('./Data/california_cities.csv')
cities.head()
cities['area_land_km2'].sort_values();
```

```python
#Extract Latd and Longd 
import seaborn as sns
lat,lon = cities['latd'],cities['longd']

population, area = cities['population_total'],cities['area_total_km2']                         
sns.set()
plt.figure(figsize=(8,6));
# lat[:10],lon[:10]
#Plot using Plot API
plt.scatter(lon,
            lat,
            c= np.log10(population)
            ,cmap= 'viridis', 
            s=area,
            alpha=0.5,
            linewidths=0);
plt.gca().set_aspect('equal');
plt.xlabel('longtitude');
plt.ylabel('latitude');
plt.colorbar(label='log$_{10}$(population)');
plt.clim(3,7);


#Create  a lenged for cities's sizes
area_range = [50,100,300,500]

for area in area_range:
 plt.scatter([],[], c='k',alpha=0.4,
             s = area,label = str(area)+ 'km$^2$')

plt.legend(labelspacing=1,title='City Area');
plt.title('City Latitude and Longitude of California');
```
